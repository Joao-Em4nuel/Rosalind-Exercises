```python
#Computing GC Content
#id:GC

#Open DNA sequence from txt file and store it in "dna" variable
dna = open("rosalind_gc.txt", "r")

#Read total DNA sequence from txt file and store it as a string in "dna_str" variable
dna_str = str(dna.read())

#Verify DNA sequence
dna_str
```

```python
#Split the fasta string into a list of strings by the character '>'
x = dna_str.split('>')
```

```python
#Filter possible empty strings from the list
x = list(filter(None, x))

#Split each string in the list by the first line break that separates the id from the DNA seq
#and save them in a list of lists
y = []
for w in x:
    y.append(w.split('\n', 1))
```

```python
#Remove all the lines breaks to concatenate the DNA sequences and save them to a list
j = []
for i in y:
    for w in i:
        j.append(w.replace('\n', ''))
```

```python
#Transform the list into a dict where each key is the DNA seq id
def Convert(a):
    it = iter(a)
    res_dct = dict(zip(it, it))
    return res_dct

j = Convert(j)
```

```python
#Transform each key value (item) to a list to associate more values and add the CG Content 
for i in j:
    j[i] = j[i].split()
    j[i].append((sum(map(j[i][0].count, ['C','G']))/len(j[i][0]))*100)
```

```python
#Figure de maximum value of CG content
m = max([list(j.values())[i][1] for i in range(3)])
```

```python
#Search for the key with the maximum value
[(i, j[i][1]) for i in j if j[i][1] == m]
```

```python
#Counting Point Mutations (Hamming distance)
#id:HAMM 

#Open DNA sequence from txt file and store it in "dna" variable
dna = open("rosalind_hamm.txt", "r")

#Read total DNA sequence from txt file and store it as a string in "dna_str" variable
dna_str = str(dna.read())

#Verify DNA sequence
dna_str
```

```python
#Split sequences by the line breaks and clean the extra line break
x = dna_str.split('\n', 1)
for i in range(len(x)):
    x[i] = x[i].replace('\n', '')
```

```python
#Save de sequences
seq1 = x[0]
seq2 = x[1]
```

```python
#Create an iterator with the string sequences and compare each character and count the dissimilar
len([i for i in zip(seq1, seq2) if i[0] != i[1]])
```
