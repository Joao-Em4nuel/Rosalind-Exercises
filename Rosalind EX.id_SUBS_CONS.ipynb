```python
#Finding a Motif in DNA
#id:SUBS

#Open DNA sequence from txt file and store it in "DNA_seq_file" variable
DNA_seq_file = open(".txt", "r")

#Read total DNA sequence and substring from txt file and store it as a string in "DNA_seq_file" variable
DNA_seq_file = str(DNA_seq_file.read()).strip('\n')

#Split and store main string and substring in diferent variables "DNA_seq" and "DNA_sub_seq", respectively
DNA_seq = DNA_seq_file.split('\n')[0]
DNA_sub_seq = DNA_seq_file.split('\n')[1]
```

```python
#Locate all positions where Motif begin

x = 0
location = []
while x < len(DNA_seq):
    #print(''.join(list(DNA_seq)[x:x + 4]))
    if DNA_sub_seq == ''.join(list(DNA_seq)[x:x + len(DNA_sub_seq)]):
        location.append(x + 1)
    x = x + 1

' '.join(str(x) for x in location)
```

```python
#Consensus and Profile
#id:CONS

#Open DNA sequence from txt file and store it in "DNA_seq_file" variable
DNA_seq_file = open("rosalind_cons.txt", "r")

#Read total DNA sequence and substring from txt file and store it as a string in "DNA_seq_file" variable
DNA_seq_file = str(DNA_seq_file.read()).strip('\n')
```

```python
#Split the fasta string into a list of strings by the character '>'
x = DNA_seq_file.split('>')
```

```python
x
```

```python
#Filter possible empty strings from the list
x = list(filter(None, x))

#Split each string in the list by the first line break that separates the id from the DNA seq
#and save them in a list of lists
y = []
for w in x:
    y.append(w.split('\n', 1))
```

```python
#Remove all the lines breaks to concatenate the DNA sequences and save them to a list
j = []
for i in y:
    for w in i:
        j.append(w.replace('\n', ''))
```

```python
#Transform the list into a dict where each key is the DNA seq id
def Convert(a):
    it = iter(a)
    res_dct = dict(zip(it, it))
    return res_dct

j = Convert(j)
```

```python
#Create a dictionary to store the profile matrix where each key is a nucleotide and a row
Profile_Matrix = {'A': [], 'C': [], 'G':[], 'T':[]}
```

```python
#Store the count of each base in each column in the profile matrix dictionary
i = 0
while i < len(j[list(j.keys())[0]]):
    for c in list(Profile_Matrix.keys()):
            Profile_Matrix[c].append(0)
    for n in list(j.keys()):
        if j[n][i] == 'A':
            Profile_Matrix['A'][i] += 1
        if j[n][i] == 'C':
            Profile_Matrix['C'][i] += 1
        if j[n][i] == 'G':
            Profile_Matrix['G'][i] += 1
        if j[n][i] == 'T':
            Profile_Matrix['T'][i] += 1
    i = i + 1
```

```python
Profile_Matrix
```

```python
i = 0
consensus = []
while i < len(Profile_Matrix[list(Profile_Matrix.keys())[0]]):
    order = []
    for c in list(Profile_Matrix.keys()):
        order.append(Profile_Matrix[c][i])
    largest = max(order)
    consensus.append([])
    for c in list(Profile_Matrix.keys()):
        if Profile_Matrix[c][i] == largest:
            consensus[i].append(c)
    i = i + 1
```

```python
for i in consensus:
    print(i[0], end = '')
print()
for c in list(Profile_Matrix.keys()):
    print(c + ': ', end = '')
    for x in Profile_Matrix[c]:
        print(x, end = ' ')
    print()
```
